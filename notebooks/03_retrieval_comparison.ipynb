# Retrieval Comparison (Phase 4)

Compare baseline naive RAG and standard KG2RAG experiment outputs.

```python
from pathlib import Path

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ROOT = Path.cwd().parent
metrics_dir = ROOT / 'outputs' / 'metrics'
split = 'dev'

experiments = ['baseline_naive', 'kg2rag_standard']
rows = []
for exp in experiments:
    path = metrics_dir / f'{exp}_metrics_{split}.json'
    if not path.exists():
        print(f'Missing metrics for {exp}: {path}')
        continue
    record = pd.read_json(path, typ='series')
    rows.append({
        'experiment': exp,
        'exact_match': record.get('accuracy', {}).get('exact_match'),
        'answer_f1': record.get('accuracy', {}).get('answer_f1'),
        'support_f1': record.get('accuracy', {}).get('support_f1'),
        'mrr': record.get('retrieval', {}).get('mrr'),
        'recall@5': record.get('retrieval', {}).get('recall@5'),
    })

comparison_df = pd.DataFrame(rows)
comparison_df
```

```python
if not comparison_df.empty:
    metric_cols = ['exact_match', 'answer_f1', 'mrr', 'recall@5']
    plot_df = comparison_df.melt(id_vars=['experiment'], value_vars=metric_cols, var_name='metric', value_name='value')

    plt.figure(figsize=(10, 5))
    sns.barplot(data=plot_df, x='metric', y='value', hue='experiment')
    plt.title('Baseline vs KG2RAG Metrics')
    plt.ylim(0, 1)
    plt.tight_layout()
    plt.show()
else:
    print('No experiment metrics found yet. Run scripts/run_experiment.py first.')
```
